```python
# 문장 임베딩 추출 함수 (CLS/Mean Pooling 선택)
import numpy as np
from tqdm import tqdm

# mean pooling (padding 마스크 반영)
def mean_pooling(last_hidden_state, attention_mask):
    mask = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
    masked = last_hidden_state * mask
    summed = masked.sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    return summed / counts

@torch.no_grad()
def embed_texts(texts, tokenizer, model, device, batch_size=32, max_length=256, pooling="cls"):
    embs = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        enc = tokenizer(batch, return_tensors="pt", truncation=True, padding=True,
                        max_length=max_length).to(device)
        out = model(**enc).last_hidden_state  # [B, T, H]
        if pooling == "cls":
            rep = out[:, 0, :]                                # [CLS]
        elif pooling == "mean":
            rep = mean_pooling(out, enc["attention_mask"])
        else:
            raise ValueError("pooling must be 'cls' or 'mean'")
        embs.append(rep.cpu().numpy())
    return np.vstack(embs)

embeddings = embed_texts(texts, tokenizer, model, device, batch_size=64)
print(embeddings.shape)  # (N, 768)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.manifold import TSNE

# 1) 임베딩 행렬: (N, D)
# embeddings = ...  # 앞 단계에서 만든 SciBERT 임베딩

N, D = embeddings.shape

# 2) 정규화: L2 노멀라이즈(권장). StandardScaler는 선택.
X = normalize(embeddings)               # 행 단위 L2 정규화
# X = StandardScaler().fit_transform(embeddings)  # 대안

# 3) PCA: 샘플/피처 수에 맞춰 n_components 자동 축소, 너무 작으면 스킵
if N >= 3 and D >= 2:
    n_comp = min(100, N - 1, D)        # 핵심: n_components ≤ min(N-1, D)
    pca = PCA(n_components=n_comp, random_state=42)
    X_pca = pca.fit_transform(X)
else:
    print(f"[WARN] 샘플이 너무 적음(N={N}, D={D}) → PCA 생략하고 원본 사용")
    X_pca = X

print("X_pca shape:", X_pca.shape)

# [시각화-1] PCA 누적 분산 설명도
import matplotlib.pyplot as plt
import numpy as np

if 'pca' in globals() and hasattr(pca, "explained_variance_ratio_"):
    evr = pca.explained_variance_ratio_
    cum = np.cumsum(evr)
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, len(evr)+1), cum, marker='o')
    plt.xlabel('Number of components')
    plt.ylabel('Cumulative explained variance')
    plt.title('PCA Cumulative Explained Variance')
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("[INFO] PCA 분산 기여도를 표시할 수 없습니다.")

from sklearn.cluster import KMeans

# 4) t-SNE(시각화 전용): perplexity < N 조건 충족, 너무 작으면 스킵
if N >= 5:
    perp = min(30, N - 1)               # t-SNE 규칙: perplexity < N
    tsne = TSNE(n_components=2, perplexity=perp, learning_rate=500,
                max_iter=1500, init="pca", random_state=42)
    X_tsne = tsne.fit_transform(X_pca)   # 또는 X
    print("X_tsne shape:", X_tsne.shape)

    # [시각화-2] t-SNE 산점도 (k-means 색상)
    if X_tsne is not None:
        # 시각화를 위해 임시로 k를 정합니다: 이미 계산된 best가 있으면 사용, 없으면 대안(예: 5)
        k_vis = best[0] if ('best' in globals() and best is not None) else min(5, max(2, X_pca.shape[0]-1))
        km_vis = KMeans(n_clusters=k_vis, n_init=10, random_state=42).fit(X_pca)
        labels_vis = km_vis.labels_
    
        plt.figure(figsize=(7, 6))
        sc = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_vis, s=40)
        plt.title(f"t-SNE colored by k-means (k={k_vis})")
        plt.xlabel("t-SNE 1"); plt.ylabel("t-SNE 2")
        plt.colorbar(sc, label='Cluster ID')
        plt.grid(True, alpha=0.2)
        plt.show()
    else:
        print("[INFO] t-SNE 좌표가 없어 산점도를 표시하지 않습니다.")

else:
    print(f"[WARN] 샘플이 너무 적음(N={N}) → t-SNE 시각화 생략")
    X_tsne = None
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import hdbscan

N = X_pca.shape[0]

# 1) k 후보를 샘플 수에 맞게 제한 (2 ≤ k ≤ N-1)
k_base = [5, 10, 20, 30, 40, 50]
k_candidates = [k for k in k_base if 2 <= k <= N-1]
if not k_candidates:
    # 최소한의 탐색: 샘플이 적을 때 2..min(10, N-1)
    k_candidates = list(range(2, max(3, min(10, N-1)+1)))

best = None
for k in k_candidates:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    pred = km.fit_predict(X_pca)

    # silhouette 유효성 체크: 2 ≤ #labels < N
    n_labels = np.unique(pred).size
    if not (2 <= n_labels < N):
        print(f"k={k:2d} | silhouette=SKIP (labels={n_labels}, N={N})")
        continue

    sil = silhouette_score(X_pca, pred)  # 필요 시 metric='cosine' 권장(임베딩 L2정규화 시)
    if best is None or sil > best[1]:
        best = (k, sil)
    print(f"k={k:2d} | silhouette={sil:.3f}")

if best is None:
    print("\n[WARN] 유효한 k가 없어 silhouette을 계산하지 못했습니다.")
else:
    print(f"\nBest k by Silhouette Score: k={best[0]}, silhouette={best[1]:.3f}")

# 2) HDBSCAN (노이즈 제외 후 유효성 체크)
clusterer = hdbscan.HDBSCAN(min_cluster_size=max(5, N//20), 
                            min_samples=max(5, N//50), 
                            metric='euclidean')
pred_hdb = clusterer.fit_predict(X_pca)

mask = pred_hdb != -1  # 노이즈 제외
n_in = int(mask.sum())
labels_in = pred_hdb[mask]
n_labels_in = np.unique(labels_in).size

if n_in > 2 and 2 <= n_labels_in < n_in:
    sil_h = silhouette_score(X_pca[mask], labels_in)
    print(f"HDBSCAN | silhouette={sil_h:.3f} | clusters={n_labels_in} (+ noise)")
else:
    print(f"[WARN] HDBSCAN 유효 클러스터 부족 → "
          f"n_in={n_in}, labels_in={n_labels_in}. min_cluster_size/min_samples를 조정해보세요.")
```
